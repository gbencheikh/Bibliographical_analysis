```python
df = pd.concat([df1, df2, df3, df4, df5, df6, df7])
```

```python
# Supprimer les doublons par rapport à la colonne 'Article_title'
df_no_duplicates = df.drop_duplicates(subset=['Article_title'])

# Supprimer les lignes avec 'Article_title' vide
df = df.dropna(subset=['Article_title'])

print(f"Nombre total d'articles = {len(df)}")
print(f"Nombre total d'articles uniques = {len(df_no_duplicates)}")
```

```python
# Modifier les valeurs de la colonne 'Journal_name'

df.loc[df['Journal_name'].str.contains('Transportation Research'), 'Journal_name'] = 'Transportation Research'
```

```python
# Afficher toute la liste des articles
for index, row in df.iterrows(): 
    print(row['Journal_name'])
```

```python
import matplotlib.pyplot as plt

# Créer un histogramme du nombre d'articles par année
plt.figure(figsize=(10, 6))
df['Publication_year'].value_counts().sort_index().plot(kind='bar', color='skyblue')
plt.title('Nombre d\'articles par année')
plt.xlabel('Année')
plt.ylabel('Nombre d\'articles')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Grouper les données par année et par journal, puis compter le nombre d'articles dans chaque groupe
journal_articles_count = df.groupby(['Publication_year', 'Journal_name']).size().unstack(fill_value=0)

# Tracer un graphique
plt.figure(figsize=(12, 6))
journal_articles_count.plot(kind='bar', stacked=True)
plt.title('Nombre d\'articles par journal et par année')
plt.xlabel('Année')
plt.ylabel('Nombre d\'articles')
plt.xticks(rotation=45)
plt.legend(title='Journal', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Compter le nombre d'articles par journal
journal_counts = df['Journal_name'].value_counts()

# Sélectionner les 10 journaux les plus publiants
top_journals = journal_counts.head(5).index

# Filtrer le DataFrame pour inclure uniquement les données correspondant aux 10 journaux les plus publiants
filtered_df = df[df['Journal_name'].isin(top_journals)]

# Grouper les données par année et par journal, puis compter le nombre d'articles dans chaque groupe
journal_articles_count_by_year = filtered_df.groupby(['Publication_year', 'Journal_name']).size().unstack(fill_value=0)

# Ajouter une colonne "Autre" pour les journaux qui ne font pas partie des 10 journaux les plus publiants
other_journals = df[~df['Journal_name'].isin(top_journals)]
other_journals_count_by_year = other_journals.groupby('Publication_year').size()
journal_articles_count_by_year['Autre'] = other_journals_count_by_year
journal_articles_count_by_year.fillna(0, inplace=True)

# Tracer le graphique
plt.figure(figsize=(12, 6))
journal_articles_count_by_year.plot(kind='bar', stacked=True)
plt.title('Nombre d\'articles par journal et par année')
plt.xlabel('Année')
plt.ylabel('Nombre d\'articles')
plt.xticks(rotation=45)
plt.legend(title='Journal', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
